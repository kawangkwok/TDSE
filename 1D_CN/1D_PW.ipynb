```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
###########THIS WORKS##############
na = open('/Users/kevinkwok/Documents/NM_project/1D_CN/0pw.csv')
csv_na = csv.reader(na)
x_vn=[]
for column in csv_na:
    x_vn.append(float(column[0]))
x_vn=np.array(x_vn)

for i in range(0,7):
    nb = open('/Users/kevinkwok/Documents/NM_project/1D_CN/'+str(200*i)+'pw.csv')
    csv_nb = csv.reader(nb)
    nmag=[]
    for column in csv_nb:
        nmag.append(float(column[3]))
    nmag=np.array(nmag)
    #plt.plot(x_vn,mag_vn,color='blue')

    #plt.plot(x_vn,nmag,color='black')
    fig, axes = plt.subplots(nrows=1, ncols=1, sharex=True)
    plt.xlim([0,1])
    plt.ylim([0,0.045])
    plt.xlabel('x')
    plt.ylabel('$\Psi^*$' '$\Psi$')
    plt.plot(x_vn,nmag,color='black',label = 't = dt*'+str(200*i))
    plt.axvline(x=0.6,linestyle='--')
    plt.plot()
    plt.legend()
    plt.savefig(str(200*i)+'pw.png')

plt.show()

```
