```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
na = open('/Users/kevinkwok/Documents/NM_project/1D_Leap/0leap.csv')
csv_na = csv.reader(na)
x_vn=[]
for column in csv_na:
    x_vn.append(float(column[0]))
x_vn=np.array(x_vn)
    
for i in (0,1,3,5):
    nb = open('/Users/kevinkwok/Documents/NM_project/1D_Leap/'+str(50*i)+'leap.csv')
    csv_nb = csv.reader(nb)
    nmag=[]
    for column in csv_nb:
        nmag.append(float(column[3]))
    nmag=np.array(nmag)
    #plt.xlim([1.5,3.5])
    #plt.ylim([0,0.015])
    plt.xlabel('x')
    plt.ylabel('$\Psi^*$' '$\Psi$')
    plt.text(1.5, 0.1, 't=0')
    plt.text(2.2, 0.14, 't=dt*50')
    plt.text(2.5, 0.1, 't=dt*150')
    plt.text(2.8, 0.06, 't=dt*250')
    plt.plot(x_vn,nmag,color='black')
    plt.savefig('free_leap.png')
plt.show()



    
 
```
