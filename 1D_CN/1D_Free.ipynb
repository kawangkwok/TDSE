```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
na = open('/Users/kevinkwok/Documents/NM_project/1D_CN/0.csv')
csv_na = csv.reader(na)
x_vn=[]
for column in csv_na:
    x_vn.append(float(column[0]))
x_vn = np.array(x_vn)

for i in (0,1,3,5):
    nb = open('/Users/kevinkwok/Documents/NM_project/1D_CN/'+str(200*i)+'.csv')
    csv_nb = csv.reader(nb)
    nmag=[]
    for column in csv_nb:
        nmag.append(float(column[3]))
    nmag=np.array(nmag)
    plt.xlim([0,1])
    plt.ylim([0,0.015])
    plt.xlabel('x')
    plt.ylabel('$\Psi^*$' '$\Psi$')
    plt.text(0.3, 0.01, 't=0')
    plt.text(0.4, 0.0135, 't=dt*200')
    plt.text(0.5, 0.0115, 't=dt*600')
    plt.text(0.625, 0.0095, 't=dt*1000')

    plt.plot(x_vn,nmag,color='black')
    plt.savefig('free.png')


plt.show()
```
